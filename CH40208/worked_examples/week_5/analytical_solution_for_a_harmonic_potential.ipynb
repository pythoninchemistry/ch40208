## Exercise: Harmonic Potential Energy Surface

**Problem:** This exercise explores the harmonic potential energy surface for a diatomic molecule:

$$U(r) = \frac{k}{2}(r-r_0)^2$$

where $k$ is the bond force constant and $r_0$ is the equilibrium bond length. We will implement this function, plot it for H₂, and verify that the minimum occurs at $r = r_0$.

### Part 1: Define the harmonic potential function

**Problem:** Write a Python function to calculate the potential energy surface for a diatomic molecule with a harmonic bond potential. Your function should take three arguments as input ($r$, $k$, and $r_0$) and return the potential energy for the input value of $r$.

**Solution:**

```python
import numpy as np
import matplotlib.pyplot as plt

%config InlineBackend.figure_format='retina'

def harmonic_potential(r, k, r_0):
    """Calculate the harmonic potential energy.
    
    Args:
        r (float): Bond length (Å). Can be a float or np.ndarray.
        k (float): Force constant (eV Å⁻^2).
        r_0 (float): Equilibrium bond length (Å).
    
    Returns:
        Potential energy (eV). Type matches input r.
    """
    return 0.5 * k * (r - r_0) ** 2

# Test the function with a single value
r_test = 1.0  # Å
k = 36.0  # eV Å⁻²
r_0 = 0.74  # Å

U_test = harmonic_potential(r_test, k, r_0)
print(f"U({r_test} Å) = {U_test:.3f} eV")
```

**Explanation:**

The function implements the formula $U(r) = \frac{1}{2}k(r-r_0)^2$ directly. Note that:
- We use `0.5` for the factor of $\frac{1}{2}$
- The function accepts both single values and NumPy arrays for $r$, making it convenient for plotting
- At $r = 1.0$ Å (which is 0.26 Å above the equilibrium position), the potential energy is about 1.22 eV


### Part 2: Plot the potential energy surface

**Problem:** Plot the harmonic potential function for H₂ with $r_0 = 0.74$ Å and $k = 36.0$ eV Å⁻² over the range $0.38 \leq r \leq 1.1$ Å.

**Solution:**

```python
# Define parameters for H₂
k = 36.0  # eV Å^2
r_0 = 0.74  # Å

# Create array of r values
r_values = np.linspace(0.38, 1.1, 200)

# Calculate potential energy at each r value
U_values = harmonic_potential(r_values, k, r_0)

# Create the plot
plt.plot(r_values, U_values)
plt.xlabel('Bond length, $r$ (Å)')
plt.ylabel('Potential energy, $U(r)$ (eV)')
plt.title('Harmonic Potential Energy Surface for H$_2$')
plt.show()
```

**Explanation:**

The plot shows the characteristic parabolic shape of a harmonic potential. The curve is symmetric around the equilibrium position at $r_0 = 0.74$ Å, which reflects the quadratic nature of the function. As we move away from equilibrium (either increasing or decreasing $r$), the potential energy increases.

Key points to note:
- We use `np.linspace()` to create 200 evenly spaced points, giving a smooth curve
- The potential rises steeply as we compress or stretch the bond away from equilibrium
- At $r = 0.38$ Å (the left edge of our plot), the bond is compressed by 0.36 Å
- At $r = 1.1$ Å (the right edge), the bond is stretched by 0.36 Å
- Due to the symmetry, the potential energy is the same at both extremes: about 2.3 eV

### Part 3: Verify the minimum position

**Problem:** Use your function to calculate the potential energy at $r = r_0$. Add a point to your plot at $(r_0, U(r_0))$ and confirm visually that this is the minimum of the potential energy surface.

**Solution:**

```python
# Calculate potential energy at equilibrium
U_at_r0 = harmonic_potential(r_0, k, r_0)

# Create the plot with the minimum marked
plt.plot(r_values, U_values, label='$U(r)$')
plt.plot(r_0, U_at_r0, 'ro', label=f'Minimum at $r_0$ = {r_0} Å')
plt.xlabel('Bond length, $r$ (Å)')
plt.ylabel('Potential energy, $U(r)$ (eV)')
plt.title('Harmonic Potential Energy Surface for H$_2$')
plt.legend()
plt.show()
```

**Explanation:**

As expected from the analytical solution, the potential energy at $r = r_0$ is exactly zero. This makes sense because:

$$U(r_0) = \frac{k}{2}(r_0 - r_0)^2 = 0$$

The red point on the plot clearly sits at the bottom of the parabola, confirming that this is indeed the minimum of the potential energy surface. In computational chemistry, we would typically need to use numerical methods to find this minimum, but for the harmonic potential we have the advantage of knowing the analytical solution: the minimum always occurs at $r = r_0$.
