```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

from figure_formatting import figure_formatting as ff

ff.set_formatting()
ff.set_formatting({'axes.formatter.limits': (-5,5)})
```

```python
from scipy.constants import Rydberg, h, c
```

```python
def emission_wavelength(n1, n2):
    return 1/(Rydberg*(1/n1**2-1/n2**2))*1e9
```

```python
lyman = [emission_wavelength(1, n2) for n2 in range(2,11)]
balmer = [emission_wavelength(2, n2) for n2 in range(3, 11)]
paschen = [emission_wavelength(3, n2) for n2 in range(4, 11)]
```

```python
from matplotlib import rcParams

rcParams['axes.formatter.limits'] = (-5, 5)

plt.figure(figsize=(6,1))
ax1 = plt.axes(frameon=False)
plt.vlines(lyman, 0.0, 1.0, color='#F16DAB', label='Lyman')
plt.vlines(balmer, 0.0, 1.0, color='#2E99CD', label='Balmer')
plt.vlines(paschen, 0.0, 1.0, color='#4CB150', label='Paschen')
plt.yticks([])
plt.xlim(0, 2000)
plt.xlabel('wavelength / nm')
plt.legend(loc=(-0.2,0.2))
plt.savefig('H_emission_calculated.png', dpi=300)
plt.show()
```
